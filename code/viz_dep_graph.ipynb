```python
import networkx as nx
import matplotlib.pyplot as plt

def create_dependency_graph(dep_graph):
    G = nx.DiGraph()
    for edge in dep_graph:
        G.add_edge(edge[0], edge[2], label=edge[1])
    return G

def visualize_graph(G):
    pos = nx.spring_layout(G, k=0.9, iterations=50)
    plt.figure(figsize=(20, 12))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', 
            node_size=3000, font_size=8, font_weight='bold')
    
    edge_labels = nx.get_edge_attributes(G, 'label')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)
    
    plt.title("Dependency Graph", fontsize=16)
    plt.axis('off')
    plt.tight_layout()
    plt.show()

# Your dependency graph data

import json

with open ("../data/indore/en_prompt_stanza.json") as f:
    data = json.load(f)

dep_graph = data['test'][0]['dep_graph']

G = create_dependency_graph(dep_graph)
visualize_graph(G)
```
