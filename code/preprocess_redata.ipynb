```python
import os
import sys
from collections import defaultdict as ddict
import pandas as pd
import numpy as np
from tqdm import tqdm
from pprint import pprint

data_dir    = '/data/shire/data/NO-BACKUP/multilingual_KGQA/SMILER/data/'
tsv_files   =  os.listdir(data_dir)
lang_dict   =  ddict(lambda: ddict(int))
label_set   =  set()

for tsv_file in tqdm(tsv_files):
    if tsv_file.endswith('_train.tsv'):
        lang = tsv_file.split('_')[0]

        if lang == 'en-small':
            continue
        elif lang == 'en-full':
            lang = 'en'

        df   = pd.read_csv(data_dir + tsv_file, sep='\t')
        labels = list(df['label'])
        for lbl in labels:
            lang_dict[lang][lbl] += 1
            label_set.add(lbl)
```

```python
label_dict  =  ddict(list)

for lbl in list(label_set):
    label_dict['label'].append(lbl)
    for lang in lang_dict:
        label_dict[lang].append(lang_dict[lang][lbl])
        

label_df = pd.DataFrame(label_dict)
display(label_df)

rel_freqs = [50, 100, 200, 500, 1000]

```

```python
## create a relation dict

from datasets import load_dataset

ds = load_dataset("Babelscape/REDFM", "all_languages")

```

```python
relation_names = ['location',
'instance of',
'country',
'part of',
'follows',
'manufacturer',
'sport ',
'owned by',
'located in or next to body of water',
'member of',
'genre',
'author',
'headquarters location',
'capital',
'child',
'notable work',
'mouth of the watercourse',
'inception',
'named after',
'occupation',
'shares border with',
'participant',
'country of citizenship',
'founded by',
'place of birth ',
'replaces',
'cast member',
'league',
'characters',
'director',
'spouse',
'sibling']

```

```python
rel_dict = ddict(int)

for rel_data in ds['train']['relations']:
    for inst in rel_data:
        rel_dict[inst['predicate']] += 1

# arrange the relations in decreasing order of frequency
rel_dict = dict(sorted(rel_dict.items(), key=lambda item: item[1], reverse=True))
print(rel_dict)

rel_map_dict = {}

for rel_name, rel_id in zip(relation_names, rel_dict.keys()):
    rel_map_dict[int(rel_id)] = rel_name
```

```python
print(rel_map_dict)

json.dump(rel_map_dict, open('../data/redfm/relations_dict.json', 'w'), indent=4)
```
